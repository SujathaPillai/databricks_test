````sql
create function dev_lh.bronze.full_name_udf(first_name STRING, last_name STRING)
RETURNS STRING
RETURN first_name || ' ' || last_name;
````

````sql
select customer_id, dev_lh.bronze.full_name_udf(first_name, last_name) as full_name, review from dev_lh.bronze.customer_reviews;
````

````sql
CREATE OR REPLACE FUNCTION dev_lh.bronze.sentiment_analysis(review STRING)
RETURNS STRING
RETURN
   CASE
     WHEN review LIKE '%great%' OR review LIKE '%good%' or  review LIKE '%excellent%' or  review LIKE '%fantastic%'or  review LIKE '%recommend%' THEN 'Positive'
     WHEN review LIKE '%not so good%' or review LIKE '%disappointed%' or review LIKE '%terrible%' or review LIKE '%never buing%' THEN 'Negative'
     ELSE 'Neutral'
   END;
````

````sql

SELECT customer_id, dev_lh.bronze.full_name_udf(first_name, last_name) AS full_name, review, dev_lh.bronze.sentiment_analysis(review) AS sentiment
FROM dev_lh.bronze.customer_reviews;
````
