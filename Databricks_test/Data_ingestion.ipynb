https://docs.databricks.com/aws/en/sql/language-manual/functions/read_files

```python
%sql
CREATE OR REPLACE TABLE dev_lh.bronze.spotify_songs_pyspark_sql AS
SELECT *, current_timestamp as ingestion_date FROM read_files(
    '/Volumes/dev_lh/bronze/raw_csv_files/spotify_songs.csv',
    format => 'csv')
```

```python
%sql
select * from  dev_lh.bronze.spotify_songs_pyspark_sql;
```
